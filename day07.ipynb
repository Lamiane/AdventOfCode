```python
def task_2(inp):
    TOTAL_WEIGHT = 'total_weight'
    WEIGHT = 'weight'
    CHILDREN = 'children'
    VISITED = 'visited'
    
    structure = {}
    possible_roots = set()  # are those who are not children
    for line in inp.split('\n'):
        information = line.split(' ')
        name = information[0]
        if name not in structure:
            structure[name] = {}
            possible_roots.add(name)  # this program has not shown before
        structure[name][WEIGHT] = int(information[1][1:-1])  # slicing in order to remove brackets
        if len(information) > 3:
            structure[name][CHILDREN] = []
            for i in range(3, len(information)):
                child = information[i].strip(',')
                structure[name][CHILDREN].append(child)
                possible_roots.discard(child)  # child cannot be the root
                if child not in structure:
                    structure[child] = {}
    
    def traverse_and_update_weights(current):
        structure[current][VISITED] = 'Yes'
        if CHILDREN in structure[current]:
            children = structure[current][CHILDREN]
            for child in children:
                err = traverse_and_update_weights(child)
                if err is not None:
                    return err
            
            # are children's weights OK? (BTW, root cannot unbalance)
            children_weights = [structure[ch][TOTAL_WEIGHT] for ch in children]
            if len(set(children_weights)) != 1:  # weights not OK       
                mean = np.mean(children_weights)  # the values that are closer are correct
                minimal = np.min(children_weights)
                diff_min = mean - minimal
                maximal = np.max(children_weights)
                diff_max = maximal - mean
                
                if diff_max < diff_min:
                    good, bad = maximal, minimal
                else:
                    good, bad = minimal, maximal
                
                diff = good - bad
                index_of_bad = children_weights.index(bad)
                bad_child = children[index_of_bad]
                value_now = structure[bad_child][WEIGHT]
                should_be = value_now + diff
                return should_be
            
            # if everything is OK, update:
            structure[current][TOTAL_WEIGHT] = structure[current][WEIGHT] + sum(children_weights)
            
        else:
            structure[current][TOTAL_WEIGHT] = structure[current][WEIGHT]  # leaf node
            

    root =  list(possible_roots)[0]  # it only has a single element so it works
    return traverse_and_update_weights(root)
```
